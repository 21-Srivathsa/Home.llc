# Every column is correlated with the house price

```python
x = copy["Avg. Area Income"]
y = copy["Price"]
sns.scatterplot(x,y)
plt.xlabel("Avg area income")
plt.ylabel("Price")
plt.title("Average area income vs Price of the house")
plt.show()
```
